## Initial analysis

### Loading and exploring the data
To begin getting a feel for the data in this file, we'll load the file into a dataframe and see what variables we have to work with. We'll also check column types, as well as some basic summary statistics. We'll make any changes that need to be made to the dataframe at this time.

```python
df = pd.read_csv("all_data.csv")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

### Modifying column names
One thing we can do right away is change the name of the column '**Life} expectancy at birth (years)**' as it's too long to work with. We'll simply rename it to **life_expectancy**.

```python
df.rename(columns={"Life expectancy at birth (years)" : "life_expectancy"}, inplace=True)
df.head()
```

### Summary statistics
While on their own, these summary statistics don't provide very specific information (such as how GDP could affect life expectancy at birth), we do get to see the mean life expectancy, mean GDP, the minimum values, and where most of the data falls. We can see that the minimum value is likely an outlier, as the IQR (the difference between Q1 and Q3) is very small.

By comparison, the IQR for GDP is slightly larger. It could indicate that after a certain GDP, there's no real tangible benefit to life expectancy. In the next section we'll plot the data to visualize it more clearly and see if this could be true.


### What other data can be identified?
It seems like this data includes countries and years, in order to see exactly which countries and years are included in our data. We can create lists that will show the unique values in the dataset for those variables

```python
countries = df.Country.unique()
years = df.Year.unique()
print(countries)
print(years)
```

It's clear that the data in this set includes data for the years 2000-2015, and includes the following countries:

- Chile
- China
- Germany
- Mexico
- United States of America

We'll be analyzing GDP and life expectancy data for these six countries.

## Visualizing the Data

### Comparison between countries in GDP and life expectancy
To visualize this, we can use a bar graph in `seaborn` which aggregates data very easily, allowing us to compare the GDP and life expectancy in these countries to each other. The y-axis has been set to a log scale, in order to be able to visualize Zimbabwe's GDP. Without this scale, it would not be visible in comparison to these other countries. Additionally, Chile would also be very difficult to see and compare. 

```python
plt.figure(figsize=(12,8))
ax2 = sns.barplot(data=df, x='Country', y='GDP')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
ax2.set_yscale('log')
ax2.set_title("GDP by Country over time")
ax2.set_ylabel('GDP in Trillions of U.S. Dollars')
```

```python
plt.figure(figsize=(12,8))
sns.set_context("notebook")
ax = sns.barplot(data=df, x='Country', y='life_expectancy')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title("Life expectancy by Country over time")
ax.set_ylabel('Life expectancy at birth (years)')
```

### Analysis
The US has the highest GDP of all the countries listed for this time period, however, it does not have the highest life expectancy. As for life expectancy, Zimbabwe finds itself in the lowest rank, with an average of about 50 years. 

## Is GDP and life expectancy increasing overall?
Line graphs can help us visualize these variables over time. We will create a `FacetGrid` using `seaborn` in order to group these subplots together. 

```python
graphGDP = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', sharey=False, sharex=False)
graphGDP = (graphGDP.map(sns.lineplot, "Year", "GDP")).add_legend()
graphGDP.set_axis_labels("Year", "GDP in Trillions of U.S. Dollars")
```

```python
graph_expectancy = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', sharey=False, sharex=False)
graph_expectancy = (graph_expectancy.map(sns.lineplot, "Year", "life_expectancy")).add_legend()
graph_expectancy.set_axis_labels("Year", "Life expectancy at birth (years)")
```

### Analysis
We can visualize a very clear increase in both GDP and life expectancy during the past 15 years. We can see more clearly how Zimbabwe's life expectancy at birth goes from a little over 45 too over 60 during this period of time. Most countries present changes of 1-4 years in life expectancy at birth, but Zimbabwe has seen the largest increase.

### Is there a correlation between GDP and life expectancy?

The simplest way to try to find a relationship between two quantitative variables is using a scatter plot. Here, should we see an upward trend, we can infer a positive correlation between the two variables. 

```python
graph_comparison = sns.FacetGrid(df,col='Country', col_wrap=3, hue='Country', sharey=False, sharex=False)
graph_comparison = (graph_comparison.map(sns.scatterplot, "GDP", "life_expectancy")).add_legend()
graph_comparison.set_axis_labels("GDP", "Life expectancy at birth (years)")
```

### Analysis
This shows a clear positive correlation. We can see that as GDP increases, life expectancy increases, as well.



## Conclusion

In conclusion, we can clearly see that GDP and life expectancy at birth are both increasing over time. There is also a clearly visible positive correlation between these variables. As GDP increases, so does life expectancy. Mexico's life expectancy presented the smallest change over this time period, with Zimbabwe presenting the largest. 

As for GDP, we have observed that although every country's GDP has been increasing, China's GDP does not show significant decreases or 'dips' over time. Other countries will have years in which their GDP decreases before recovering and continuing to increase, whereas China's GDP has been increasingly constantly over the period of time in this dataset.
